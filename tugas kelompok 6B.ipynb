```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Definisikan variabel
x, y = sp.symbols('x y')

# Definisikan fungsi
f = x**2 + y**2

# Hitung turunan pertama
fx = sp.diff(f, x)
fy = sp.diff(f, y)

# Cari titik stasioner
stationary_points = sp.solve((fx, fy), (x, y))

# Hitung turunan kedua
fxx = sp.diff(fx, x)
fyy = sp.diff(fy, y)
fxy = sp.diff(fx, y)

# Tentukan sifat titik stasioner
D = fxx * fyy - fxy**2
nature = "Minima" if D.subs(stationary_points) > 0 and fxx.subs(stationary_points) > 0 else "Saddle Point" if D.subs(stationary_points) < 0 else "Maxima"

# Cetak hasil
print("Titik Stasioner:", stationary_points)
print("Sifat Titik Stasioner:", nature)

# Buat peta kontur
X = np.linspace(-2, 2, 100)
Y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**2

plt.contour(X, Y, Z, levels=20)
plt.scatter([0], [0], color='red', label='Stationary Point')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Contour Map of f(x, y) = x² + y²')
plt.legend()
plt.show()
```

```python
import sympy as sp

x, y = sp.symbols('x y')

z = x**2 + y**2

dz_dx = sp.diff(z, x)
dz_dy = sp.diff(z, y)

stationary_points = sp.solve([dz_dx, dz_dy], (x, y))
print("Titik stasioner:", stationary_points)

if stationary_points:
    stationary_point = (stationary_points[x], stationary_points[y])
    print("Koordinat titik stasioner:", stationary_point)

    d2z_dx2 = sp.diff(dz_dx, x)
    d2z_dy2 = sp.diff(dz_dy, y)
    d2z_dxdy = sp.diff(dz_dx, y)

    d2z_dx2_value = d2z_dx2.subs({x: stationary_point[0], y: stationary_point[1]})
    d2z_dy2_value = d2z_dy2.subs({x: stationary_point[0], y: stationary_point[1]})
    d2z_dxdy_value = d2z_dxdy.subs({x: stationary_point[0], y: stationary_point[1]})

    print("Turunan kedua di titik stasioner:")
    print("∂²z/∂x² =", d2z_dx2_value)
    print("∂²z/∂y² =", d2z_dy2_value)
    print("∂²z/∂x∂y =", d2z_dxdy_value)

    delta = d2z_dx2_value * d2z_dy2_value - d2z_dxdy_value**2
    print("Δ =", delta)

    if delta < 0:
        print("Karena Δ < 0, titik (0, 0) adalah titik saddle.")
    elif d2z_dx2_value > 0:
        print("Karena Δ > 0 dan ∂²z/∂x² > 0, titik (0, 0) adalah titik minimum.")
    else:
        print("Titik (0, 0) adalah titik maksimum.")

    sp.plotting.plot_implicit(sp.Eq(z, 1), (x, -3, 3), (y, -3, 3), title='Peta Kontur dari z = x^2 + y^2', xlabel='x', ylabel='y')
else:
    print("Tidak ada titik stasioner ditemukan.")
```

```python
import sympy as sp

x, y = sp.symbols('x y')

f = 2*x*y**2 - x*y + 4*x + 8

f_x = sp.diff(f, x)
f_y = sp.diff(f, y)

titik_kritis = sp.solve((f_x, f_y), (x, y))

f_xx = sp.diff(f_x, x)
f_yy = sp.diff(f_y, y)
f_xy = sp.diff(f_x, y)

print("Titik Kritis:")
for titik in titik_kritis:
    print(f"x = {titik[0]}, y = {titik[1]}")

print("\nTurunan Pertama:")
print(f"∂f/∂x = {f_x}")
print(f"∂f/∂y = {f_y}")

print("\nTurunan Kedua:")
print(f"∂²f/∂x² = {f_xx}")
print(f"∂²f/∂y² = {f_yy}")
print(f"∂²f/∂x∂y = {f_xy}")

Hessian = sp.Matrix([[f_xx, f_xy], [f_xy, f_yy]])

for titik in titik_kritis:
    H_at_titik = Hessian.subs({x: titik[0], y: titik[1]})
    det_H_at_titik = H_at_titik.det()
    
    print(f"\nDeterminant Hessian di titik {titik}: Δ = {det_H_at_titik}")
    
    if det_H_at_titik > 0:
        if f_xx.subs({x: titik[0], y: titik[1]}) > 0:
            print(f"Titik {titik} adalah titik minimum.")
        else:
            print(f"Titik {titik} adalah titik maksimum.")
    elif det_H_at_titik < 0:
        print(f"Titik {titik} adalah titik saddle.")
    else:
        print(f"Titik {titik} tidak dapat disimpulkan.")
```

```python
import sympy as sp

x, y = sp.symbols('x y')

f = x * y**2 - 2 * x + 4 * y + 8

f_x = sp.diff(f, x)
f_y = sp.diff(f, y)

titik_kritis = sp.solve((f_x, f_y), (x, y))

f_xx = sp.diff(f_x, x)
f_yy = sp.diff(f_y, y)
f_xy = sp.diff(f_x, y)

hasil = []
for titik in titik_kritis:
    D = f_xx.subs({x: titik[0], y: titik[1]}) * f_yy.subs({x: titik[0], y: titik[1]}) - (f_xy.subs({x: titik[0], y: titik[1]})**2)
    f_yy_value = f_yy.subs({x: titik[0], y: titik[1]})
    
    if D > 0:
        if f_yy_value > 0:
            hasil.append((titik, 'Minimum Lokal'))
        else:
            hasil.append((titik, 'Maksimum Lokal'))
    elif D < 0:
        hasil.append((titik, 'Titik Saddle'))
    else:
        hasil.append((titik, 'Tidak Dapat Disimpulkan'))

for result in hasil:
    print(f"Titik: {result[0]}, Klasifikasi: {result[1]}")
```

```python
import sympy as sp

x, y = sp.symbols('x y')

z = -2*y - 4*x  

f_x = sp.diff(z, x)
f_y = sp.diff(z, y)

print("Turunan Pertama:")
print(f"∂z/∂x = {f_x}")
print(f"∂z/∂y = {f_y}")

titik_kritis = sp.solve((f_x, f_y), (x, y))

print("\nTitik Kritis:")
print(titik_kritis)

f_xx = sp.diff(f_x, x)
f_yy = sp.diff(f_y, y)
f_xy = sp.diff(f_x, y)

print("\nTurunan Kedua:")
print(f"∂²z/∂x² = {f_xx}")
print(f"∂²z/∂y² = {f_yy}")
print(f"∂²z/∂x∂y = {f_xy}")

H = sp.Matrix([[f_xx, f_xy], [f_xy, f_yy]])

hasil = []
for titik in titik_kritis:
    if isinstance(titik, tuple):
        H_di_titik = H.subs({x: titik[0], y: titik[1]})
    else:
        H_di_titik = H.subs({x: titik[0], y: titik[1]})

    determinan = H_di_titik.det()
    
    print(f"\nDeterminant Hessian di titik {titik}: {determinant}")

    if determinan > 0:
        if f_xx.subs({x: titik[0], y: titik[1]}) > 0:
            hasil.append((titik, 'Minimum Lokal'))
        else:
            hasil.append((titik, 'Maksimum Lokal'))
    elif determinan < 0:
        hasil.append((titik, 'Titik Saddle'))
    else:
        hasil.append((titik, 'Tidak Dapat Disimpulkan'))

print("\nHasil:")
for result in hasil:
    print(f"Titik: {result[0]}, Sifat: {result[1]}")
```

```python
import sympy as sp

# Definisikan variabel
x, y = sp.symbols('x y')

# Definisikan fungsi
f = 2*x + 2*y - 2*x*y - 2*x**2 - y**2 + 4

# Hitung turunan pertama
fx = sp.diff(f, x)
fy = sp.diff(f, y)

# Cari titik kritis
critical_points = sp.solve([fx, fy], (x, y), dict=True)

# Hitung turunan kedua
fxx = sp.diff(fx, x)
fyy = sp.diff(fy, y)
fxy = sp.diff(fx, y)

# Hitung Determinan Hessian
D = fxx * fyy - fxy**2

# Evaluasi hasil untuk setiap titik kritis
for point in critical_points:
    x_val = point[x]
    y_val = point[y]
    hessian_value = D.subs({x: x_val, y: y_val})
    fxx_value = fxx.subs({x: x_val, y: y_val})
    f_value = f.subs({x: x_val, y: y_val})
    
    if hessian_value > 0:
        if fxx_value > 0:
            result = "Minimum lokal"
        else:
            result = "Maksimum lokal"
    elif hessian_value < 0:
        result = "Titik pelana"
    else:
        result = "Tidak dapat ditentukan"
    
    print(f"Titik kritis: ({x_val}, {y_val}), Jenis: {result}, Nilai fungsi: {f_value}")
```
